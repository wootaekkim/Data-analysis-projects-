We use lable encoding because label encoding works well with tree based models, and this is a way to transform categorial variable into numerical variable withour increasing number of features which can possibly make the model's prediction performace worse

### Modeling

first split between training and testing set

```python
X = df.drop(['Price'], axis = 1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Since all the numerical variables are far from normal distribution, choose tree based model(LightGBM, XGBoost) to trarin them

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```
